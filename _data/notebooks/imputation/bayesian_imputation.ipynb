```r
library(ca)
library(HydeNet)
library(ggplot2)
```

```r
data("wg93")
```

```r
seed <- 42
set.seed(seed)
theme_set(theme_minimal())
```

```r
head(wg93)
```

```r
t <- cpt(A ~ B + age, wg93)
```

```r
likelihood <- t[B = 3, age=1, ]
```

```r
prior <- t[A = c(1:5)]
```

```r
post <- c()
for(i in seq(1:5)) {
    post <- c(post, (likelihood[[i]] + 1e-9) * (prior[[i]] + 1e-9))
}
```

```r
post <- post / sum(post)
```

```r
barplot(prior, names.arg = c("1", "2", "3", "4", "5"), main = "Prior - P(A)")
```

```r
barplot(likelihood, names.arg = c("1", "2", "3", "4", "5"), main = "Likelihood - P(A | B = 3, age = 1)")
```

```r
barplot(post, names.arg = c("1", "2", "3", "4", "5"), main = "Posterior - P(A) * P(A | B = 3, age = 1)")
```

```r
prediction <- which.max(post)
prediction
```
